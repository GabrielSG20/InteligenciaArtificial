```python
# Importação de bibliotecas necessárias para análise e visualização de dados, além de aprendizado de máquina.
import matplotlib.pyplot as plt # Para criação de gráficos.
import numpy as np # Lib para operações matemáticas complexas e manipulação de arrays.
from sklearn.linear_model import LinearRegression # Para criação do modelo de regressão linear;
from sklearn.metrics import mean_squared_error, r2_score # Para calcular métricas de avaliação de modelo.

# Dados fornecidos pelo usuário, onde os gastos com publicidade estão em milhares de reais e as vendas em unidades.
gastos_publicidade = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5, 6.0, 6.5, 7.0]
licencas_vendidas = [50, 55, 60, 68, 72, 80, 85, 88, 90, 95, 100, 105]
```

```python
# Conversão de listas em arrays numpy, formato necessário para uso em scikit-learn e ajuste para valores em reais.
x = np.array(gastos_publicidade).reshape(-1, 1) * 1000 # Gastos com publicidade em Real
y = np.array(licencas_vendidas) # Número de licenças vendidas.

# Criação do modelo de regressão linear utilizando a biblioteca scikit-learn.
modelo = LinearRegression()
modelo.fit(x, y) # Treinamento do modelo.

y_pred = modelo.predict(x) # Predição dos valores de y com base nos valores de x.
mse = mean_squared_error(y, y_pred) # Cálculo do erro quadrático médio.
r2 = r2_score(y, y_pred) # Cálculo do coeficiente de determinação.

# Geração e configuração do gráfico para visualizar os dados e a linha de regressão.
plt.figure(figsize=(10, 6)) # Definição do tamanho da figura do gráfico.
plt.scatter(x, y, color='blue', label='Dados reais') # Gráfico de regressão com dados reais.
plt.plot(x, y_pred, color='red', label='Linha de regressão') # Linha de regressão.

# Adição de títulos e rótulos aos eixos do gráfico para melhor compreensão.
plt.title('Regressão Linear - Vendas de Licenças vs. Gastos com Publicidade')
plt.xlabel('Gastos com Publicidade (R$)')
plt.ylabel('Licenças Vendidas')
plt.legend() # Inclusão de legenda no gráfico.
plt.show() # Exibição no gráfico.
```

```python
# Uso do modelo treinado para prever o número de licenças vendidas com um gasto de R$8.000,00 em publicidade.
previsao = modelo.predict(np.array([[8000]]))

# Os resultados das métricas de avaliação e a previsão são apresentados.
mse, r2, previsao
```
